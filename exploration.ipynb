```python
# imports

import pandas as pd
import ast
```

## Cleaning Data
The database was collected in the official Stardew Valley wiki manually inputted into an Excel file and converted into a CSV file.


```python
# Reading and Locating Paths

file_path = "data/Marriageable Villagers.csv"
df = pd.read_csv(file_path, index_col = 0)
```

```python
df.head()
df.dtypes
df.info()
```

```python
# Setting Max Displays

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)

display(df)
```

The output shows that the gift and movie columns contain string representations of lists
and will need to be converted before analysis.

With that using ast.literal_eval is needed
to safely convert them into actual Python lists.

```python
# Converting Columns into Lists

df["Specific Loved Gifts"] = df["Specific Loved Gifts"].apply(ast.literal_eval)
df["Specific Liked Gifts"] = df["Specific Liked Gifts"].apply(ast.literal_eval)
df["Specific Disliked Gifts"] = df["Specific Disliked Gifts"].apply(ast.literal_eval)
df["Specific Hated Gifts"] = df["Specific Hated Gifts"].apply(ast.literal_eval)
df["Loved Movies"] = df["Loved Movies"].apply(ast.literal_eval)
```

Concatenating the gift columns into one master columns for cleaning verification

```python
all_gifts = pd.concat([
    df["Specific Loved Gifts"],
    df["Specific Liked Gifts"],
    df["Specific Disliked Gifts"],
    df["Specific Hated Gifts"]
])

print(df["Loved Movies"].explode().unique().tolist())
print()
print(all_gifts.explode().unique().tolist())
```

### Exploratory Data Analysis

```python
cat_counts = {}
for category in gift_categories:
    count = 0
    for _, row in df.iterrows():
        if set(row['Specific Loved Gifts']) & set(gift_categories[category]):
            count += 1
    cat_counts[category] = count

print(cat_counts)
```
